This notebook is simply used to better understand the taxi domain environment

```python
import gymnasium as gym
```

```python
env = gym.make('Taxi-v3', render_mode="rgb_array")
step_0 = env.reset()
print(step_0)
```

```python
env.spec
```

```python
print(f"State space: {env.observation_space}, of size: {env.observation_space.n}")
print(f"Action space: {env.action_space}, of size {env.action_space.n}")
```

```python
actions = ["S", "N", "E", "W", "PICK_UP", "DROP"]
```

----

Plot the environment

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(env.render())
```

```python
a, b, c, d = env.decode(step_0[0])
```

```python
print(a, b, c, d)
```

Sample Transition

```python
import random
```

```python
random_action = random.randint(0, 4)
print(f"Action taken: {actions[random_action]}")
```

```python
next_state, reward, done, truncated, info = env.step(random_action)
```

```python
print (f"Next state: {next_state}")
print (f"Reward recieved: {reward}")
print (f"Terminal state: {done}")
print (f"Truncated: {truncated}")
print (f"Info: {info}")
```
